```python

import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
df1 = DataFrame({'süt1':np.arange(6),'ortak':['a','b','c','a','b','c']})
df1
```

```python
df2 = DataFrame({'süt2':np.arange(4),'ortak':['a','a','m','b']})
df2
```

```python
pd.merge(df1,df2)#ortak degerleri birleştirdi.
```

```python
pd.merge(df1,df2,on='ortak')#hangi eksene göre birleştirme ekseni
```

```python
pd.merge(df1,df2,on='ortak',how='left')#leftteki ortak olmayan degerleri de tut.
```

```python
pd.merge(df1,df2,on='ortak',how='right')
```

```python
pd.merge(df1,df2,on='ortak',how='inner')#default deger
```

```python
pd.merge(df1,df2,on='ortak',how='outer')#komple hepsini alıyor.
```

```python
pd.merge(df1,df2,left_on='süt1',right_on='süt2')#ortak olan degeri 2 ye böldü
```
