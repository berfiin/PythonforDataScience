```python
import numpy as np
import matplotlib.pyplot as plt


%matplotlib inline #her grafiği ekranda görebilmek için.
```

```python
x = np.linspace(0,5,21)
x #iki tane aralıgı eşit parçalara böldü.
```

```python
y1 = 2**x
y2 = 5**x
```

```python
plt.plot(x,y1)
plt.xlabel('Emek',fontsize=15)
plt.ylabel('Ekmek',fontsize=15)
plt.title('Hayat',fontsize=15)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,'r')

plt.subplot(1,2,2)
plt.plot(x,y2,'b')

plt.tight_layout() #aradaki boşluk için
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_xlabel('berfin')
axes.plot(x,y1)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.12,0.6,0.4,0.3])

axes1.plot(x,y2)
axes.plot(x,y1)
```
