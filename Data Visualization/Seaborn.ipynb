```python
import pandas as pd
import numpy as np

from pandas import DataFrame

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
bahsis_data = pd.read_csv('veri_bahsis.csv', sep='\t')
bahsis_data
```

```python
sns.distplot(bahsis_data['hesap'], kde=True, bins=10, rug=True) 
sns.set_style('darkgrid')#olasılık dagılımları
```

```python
fig, axes = plt.subplots(figsize=(8,2))
sns.distplot(bahsis_data['hesap'], bins=15, rug=True, ax=axes) 
```

```python
sns.jointplot(x='bahsis', y='hesap', data=bahsis_data, kind='kde')
```

```python
#yorum daha çok 10 la 20 tl hesap ödeyen insanlar en çok bahsisi veren insanlar
```

* pairplot = numeric deg. olan sutunların birbirlerine göre dagılımlarını gösterir.

```python
bahsis_data.head()
```

```python
sns.pairplot(data=bahsis_data, diag_kind='kde',kind='reg', hue='cinsiyet')
```

```python
bahsis_data.head()
```

* barplot = kategorize edilen degerleri çizdirir

```python
sns.barplot(x='gün', y='hesap',data=bahsis_data)
```

```python
sns.barplot(x='hesap', y='gün', data=bahsis_data,estimator=np.max)
```

```python
sns.countplot(x='sigara',data=bahsis_data,hue='cinsiyet')
```

* boxplot = istatiksel olarak bilgi

```python
sns.boxplot(x='gün', y='hesap', data=bahsis_data)
```

```python
sns.boxplot(y='zaman',x='hesap',data=bahsis_data,hue='cinsiyet')
```

```python
sns.violinplot(x='gün', y='bahsis',data=bahsis_data)
```

```python
sns.violinplot(x='gün', y='bahsis', data=bahsis_data, hue='cinsiyet',split=True)
```

* stripplot=bir verinin hem dagılımı hem degeri

```python
sns.stripplot(x='zaman',y='hesap',data=bahsis_data)
```

```python
sns.stripplot(x='hesap', y='zaman',data=bahsis_data,hue='sigara')
```

* swarmplot = her noktayı o  kategori için belli bir genişliğe yayarak,hiçbir nokta üst üstü gelmeyecek şekilde çizdiren grafik tipidir.


```python
sns.swarmplot(y='hesap',x='zaman',data=bahsis_data)
```

```python
sns.swarmplot(y='hesap',x='zaman',data=bahsis_data,color='black')
sns.violinplot(y='hesap',x='zaman',data=bahsis_data)
```

* catplot=her kategoriye göre en geniş kapsam 

```python
sns.catplot(x='gün',y='hesap',data=bahsis_data,kind='boxen',hue='sigara')#kind=bar,strip ,boxen vs olabilir.
```

```python
sns.catplot(x='gün',y='hesap',data=bahsis_data,kind='point',col='sigara')
```
